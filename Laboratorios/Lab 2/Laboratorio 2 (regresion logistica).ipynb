Funcion de la Sigmoide

```python
def calcularSigmoide(z):
    # Calcula la sigmoide de una entrada z
    # convierte la intrada a un arreglo numpy
    z = np.array(z)

    g = np.zeros(z.shape)

    g = 1 / (1 + np.exp(-z))

    return g
```

```python
# Prueba la implementacion de la funcion sigmoid
z = [0, 0.5, 1]
g = calcularSigmoide(z)

print('g(', z, ') = ', g)
```

Funcion Costo Gradiente

agregamos la una primera columna de unos

```python
# Configurar la matriz adecuadamente, y agregar una columna de unos que corresponde al termino de intercepci칩n.
m, n = X.shape
# Agraga el termino de intercepci칩n a A
X = np.concatenate([np.ones((m, 1)), X], axis=1)
```

Funcion de costo en una regresion logistica

```python
def calcularCosto(theta, X, y):
    # Inicializar algunos valores utiles
    m = y.size  # numero de ejemplos de entrenamiento

    J = 0
    h = calcularSigmoide(X.dot(theta.T))
    J = (1 / m) * np.sum(-y.dot(np.log(h)) - (1 - y).dot(np.log(1 - h)))

    return J
```

```python
def descensoGradiente(theta, X, y, alpha, num_iters):
    # Inicializa algunos valores
    m = y.shape[0] # numero de ejemplos de entrenamiento

    # realiza una copia de theta, el cual ser치 acutalizada por el descenso por el gradiente
    theta = theta.copy()
    J_history = []

    for i in range(num_iters):
        h = calcularSigmoide(X.dot(theta.T))
        theta = theta - (alpha / m) * (h - y).dot(X)

        J_history.append(calcularCosto(theta, X, y))
    return theta, J_history
```

Funcion para calcular la precision de las predicciones

```python
def calcularPrecision(predictions, labels):
    # Convertir predicciones a etiquetas binarias (0 o 1)
    predictions_binary = (predictions >= 0.5).astype(int)

    # Calcular la precisi칩n
    correct_predictions = np.sum(predictions_binary == labels)
    total_examples = len(labels)
    precision = correct_predictions / total_examples

    return precision
```

Separar el 80% de los datos para entrenamiento y el 20% para hacer las pruebas

```python
from sklearn.model_selection import train_test_split
# Dividir los datos en conjuntos de entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.2, random_state=42)
```

Configuracion de Hiperparametros para el entrenamiento del Modelo

```python
# Configurar hiperpar치metros y realizar descenso por el gradiente en el conjunto de entrenamiento
alpha = 0.25
num_iters = 1000       #con 500 iteraciones ya tenemos una buena convergencia
theta = np.zeros(18)
theta, J_history = descensoGradiente(theta, X_train, y_train, alpha, num_iters)

# Muestra los resultados del descenso por el gradiente
print('theta calculado por el descenso por el gradiente: \n',(theta))


# Graficar la convergencia del costo en el conjunto de entrenamiento
pyplot.plot(np.arange(len(J_history)), (J_history), lw=2)

pyplot.xlabel('N칰mero de iteraciones')
pyplot.ylabel('Costo J')
pyplot.title('Convergencia del costo en el conjunto de entrenamiento')
pyplot.show()
```

Y por ultimo Hacemos las predicciones con el otro 20% de datos faltantes para la evaluacion del Modelo

```python
# Realizar predicciones en el conjunto de prueba
y_predicha = calcularSigmoide(np.dot(X_test, theta))

# Convertir predicciones a etiquetas binarias (0 o 1) usando un umbral de 0.5
predictions_binary = (y_predicha >= 0.5).astype(int)

# Calcular la cantidad de predicciones correctas e incorrectas
correct_predictions = np.sum(predictions_binary == y_test)
incorrect_predictions = len(y_test) - correct_predictions

# Calcular el porcentaje de precisi칩n en el conjunto de prueba
precision_test = correct_predictions / len(y_test) * 100

# Imprimir resultados de las pruebas en el conjunto de prueba
print('Resultados en el conjunto de prueba:')
print(f'Predicciones correctas: {y_predicha}')
print(f'Predicciones incorrectas: {incorrect_predictions}')
print(f'Porcentaje de precisi칩n: {precision_test:.2f}%')
```
